```python
import pandas as pd

sales_data = pd.read_excel('dataset.xlsx')
```

```python
sales_data.isnull().sum()
```

```python
#cleaning the data (removing the null rows)

sales_data_cleaned = sales_data.dropna(subset = ['Amount'])
```

```python
sales_data_cleaned.isnull().sum()
```

```python
big_purchases = sales_data_cleaned[sales_data_cleaned['Amount'] > 1000]
```

```python
print(big_purchases[['Order ID','Amount','ship-state']].head())
```

```python
#how many sales greater than $1000

big_purchases['Amount'].count()
```

```python
filtered_data = sales_data_cleaned[(sales_data_cleaned['Category'] == 'Tops') & (sales_data_cleaned['Qty'] == 3)]
```

```python
print(filtered_data['Order ID'].count())
```

```python
#grouping the data for aggregations

category_totals = sales_data_cleaned.groupby('Category', as_index=False)['Amount'].sum()
print(category_totals)
```

```python
#sorting this data

sorted_category_sums = category_totals.sort_values('Amount', ascending=False)
print(sorted_category_sums)
```

```python
#average of the sales by Category and Fulfilment

avg_catfulfil = sales_data_cleaned.groupby(['Category','Fulfilment'], as_index=False)['Amount'].mean()
print(avg_catfulfil)

sorted_avg_catfulfil = avg_catfulfil.sort_values('Amount', ascending=False)
```

```python
pd.set_option('display.max_columns', None)
sales_data.columns

sales_data.head(10)
```

```python
#calculate the total sales by shipment and fulfilment

total_cs_fulfil = sales_data_cleaned.groupby(['Courier Status','Fulfilment'], as_index=False)['Amount'].sum()
print(total_cs_fulfil)
```

```python
sorted_total_cs_fulfil = total_cs_fulfil.sort_values('Amount', ascending=False)
print(sorted_total_cs_fulfil)
```

```python
#Renaming the columns

sorted_total_cs_fulfil.rename(columns={'Courier Status':'Shipment'}, inplace=True)
print(sorted_total_cs_fulfil)
```

```python
#Exporting the data

sorted_total_cs_fulfil.to_excel('ShippedVsUnshipped.xlsx', index=False)
sorted_avg_catfulfil.to_excel('Average_by_category_and_Fulfilment.xlsx', index=False)
```
